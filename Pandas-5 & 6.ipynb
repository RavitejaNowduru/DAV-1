```python
df2=pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns=["A","B","C","D"])
df2
```

```python
# how to fill the NaN values with the mean value
df2.fillna(df2['A'].mean()) # It is filling the values of A row in all mssing values of df2
```

```python
df2.fillna(df2.mean()) #Here it is filling missing values with the mean of each column
#But still NaN are present
```

```python
df2.fillna(df2.mean()).fillna(0) #Here the missing values are filled with zero after filling 
#with mean values as C column is all NaN
```

```python
df2
```

```python
#If I want to fill only in A column
df2['A'].fillna(df2['A'].mean())
```

```python
#Forward Fill
df2.fillna(method ='ffill')
#Values of NaN are being replaced with its closest preceding value. Top to Bottom Approach
```

```python
#Backward Fill
df2.fillna(method ='bfill')
#Values of NaN are being replaced with its closest next value. Bottom to Top Approach
```

```python
def temp_mean(x):
    x["Temp_avg"]=np.round(x["Temperature"].mean(),2)
    return x
```

```python
df_tidy=df_tidy.groupby("Drug_Name").apply(temp_mean)
df_tidy.head(20)
```

```python
df_tidy['Temperature'].fillna(df_tidy['Temp_avg'], inplace=True)
```

```python
df_tidy.head(20)
```

```python
def pressure_mean(x):
    x['Avg_pressure'] = np.round(x['Pressure'].mean(),2)
    return x
```

```python
df_tidy=df_tidy.groupby("Drug_Name").apply(pressure_mean)
df_tidy.head(20)
```

```python
df_tidy['Pressure'].fillna(df_tidy['Avg_pressure'], inplace = True)
```

```python
temp_points=[1,20,24,27,100]
temp_labels=["low","medium","high","very_high"]
```

```python
df_tidy['Temp Category'] = pd.cut(df_tidy['Temperature'],bins = temp_points,labels = temp_labels)
```
